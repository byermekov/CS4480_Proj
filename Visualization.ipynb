```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns

%matplotlib inline
```

```python
df_positive = pd.read_csv('tfidf_positive.txt', sep='\t')
df_negative = pd.read_csv('tfidf_negative.txt', sep='\t')
```

```python
word_freq_positive = dict(zip(df_positive.word, df_positive.tfidf))

wordcloud_pos = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq_positive)

plt.figure(figsize=(15, 7.5))
plt.imshow(wordcloud_pos, interpolation='bilinear')
plt.title('Word Cloud for Positive Reviews', fontsize=20)
plt.axis('off')
plt.show()
```

```python
word_freq_negative = dict(zip(df_negative.word, df_negative.tfidf))

wordcloud_neg = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq_negative)

plt.figure(figsize=(15, 7.5))
plt.imshow(wordcloud_neg, interpolation='bilinear')
plt.title('Word Cloud for Negative Reviews', fontsize=20)
plt.axis('off')
plt.show()
```

```python
top10_positive = df_positive.sort_values(by='tfidf', ascending=False).head(10)
print("Top 10 Frequent Words in Positive Reviews:")
top10_positive
```

```python
top10_negative = df_negative.sort_values(by='tfidf', ascending=False).head(10)
print("Top 10 Frequent Words in Negative Reviews:")
top10_negative
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='tfidf', y='word', data=top10_positive, palette='Greens_d')
plt.title('Top 10 Words in Positive Reviews', fontsize=16)
plt.xlabel('TF-IDF Score')
plt.ylabel('Words')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='tfidf', y='word', data=top10_negative, palette='Reds_d')
plt.title('Top 10 Words in Negative Reviews', fontsize=16)
plt.xlabel('TF-IDF Score')
plt.ylabel('Words')
plt.show()
```
